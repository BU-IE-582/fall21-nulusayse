```r
#installing required packages
install.packages("TunePareto")
install.packages("kknn")
install.packages("data.table")
install.packages("class")
library(TunePareto)
library(kknn)
library(data.table)
library(class)
```

```r
#loading data
x_traindata <- as.data.table(read.table('uWaveGestureLibrary_X_TRAIN'))
y_traindata <- as.data.table(read.table('uWaveGestureLibrary_Y_TRAIN'))
z_traindata <- as.data.table(read.table('uWaveGestureLibrary_Z_TRAIN'))
x_testdata <- as.data.table(read.table('uWaveGestureLibrary_X_TEST'))
y_testdata <- as.data.table(read.table('uWaveGestureLibrary_Y_TEST'))
z_testdata <- as.data.table(read.table('uWaveGestureLibrary_Z_TEST'))
head(x_traindata)
```

```r
#defining column names
colnames(x_traindata)=as.character(c(1:316))
colnames(y_traindata)=as.character(c(317:632))
colnames(z_traindata)=as.character(c(633:948))
colnames(x_testdata)=as.character(c(1:316))
colnames(y_testdata)=as.character(c(317:632))
colnames(z_testdata)=as.character(c(633:948))
head(x_traindata)
```

```r
#dropping the class column
x_traindata_without_class<-x_traindata[,-1]
y_traindata_without_class<-y_traindata[,-1]
z_traindata_without_class<-z_traindata[,-1]
x_testdata_without_class<-x_testdata[,-1]
y_testdata_without_class<-y_testdata[,-1]
z_testdata_without_class<-z_testdata[,-1]
head(x_traindata_without_class)
```

```r
combinedtraindata <- cbind(x_traindata_without_class,y_traindata_without_class,z_traindata_without_class)
combinedtestdata <- cbind(x_testdata_without_class,y_testdata_without_class,z_testdata_without_class)
class_data_train <- as.matrix(x_traindata[,1])
class_data_test <- as.matrix(x_testdata[,1])
combinedtraindata <- scale(combinedtraindata)
combinedtestdata <- scale(combinedtestdata)
combinedtraindata_class<-cbind(class_data_train,combinedtraindata)
combinedtestdata_class<-cbind(class_data_test,combinedtestdata)
```

```r
#Different k levels will be tried
k_levels=c(1:10)
nofReplications=1
nFolds=10
indices=generateCVRuns(class_data_train,nofReplications,nFolds,stratified=TRUE)
```

I used two different measures: Euclidian and Manhattan. 

```r
#Euclidean
cvresult=data.table()
for(i in 1:nofReplications) {
  thisReplication=indices[[i]]
  for(j in 1:nFolds){
    testindices=thisReplication[[j]]
    
    cvtrain=combinedtraindata[-testindices,]        
    cvtest=combinedtraindata[testindices,]
    train_class = class_data_train[-testindices]
    

    for(y in 1:length(k_levels)){
      param_k=k_levels[y]
      predict_knn=knn(cvtrain, cvtest,train_class, k = param_k)
      cvresult=rbind(cvresult,data.table(Replication=i,Fold=j, Method='knn',Klev=param_k,TestId=testindices,
                                         Predictions=as.numeric(as.character(predict_knn)),Real=class_data_train[testindices]))
    }   
  }    
}
cvresult[,list(Accu=mean(Predictions==Real)),by=list(Method,Klev)]
```

We can see, when k is 3, accuracy is the highest then 3 is chosen as optimal.

```r
cvresult_manhattan=data.table()
for(i in 1:nofReplications) {
  thisReplication=indices[[i]]
  for(j in 1:nFolds){
    testindices=thisReplication[[j]]
    
    cvtrain=combinedtraindata[-testindices,]        
    cvtest=combinedtraindata[testindices,]
    train_class = class_data_train[-testindices]
    
    for(y in 1:length(k_levels)){
      param_k=k_levels[y]
      predict_knn=kknn(train_class~., as.data.frame(cvtrain), as.data.frame(cvtest),k=param_k,distance = 1,kernel = "rectangular")
      cvresult_manhattan=rbind(cvresult_manhattan,data.table(Replication=i,Fold=j,Method='knn',Klev=param_k,TestId=testindices,
                                         Predictions=as.numeric(as.character(predict_knn$fitted.values)),Real=class_data_train[testindices]))
    }   
  }    
}
cvresult_manhattan[,list(Accu=mean(Predictions==Real)),by=list(Method,Klev)]
```

We can see, when k is 1, accuracy is the highest then 3 is chosen as optimal.

```r
start_time <- Sys.time()
predict_knn_euclidian=knn(combinedtraindata, combinedtestdata,class_data_train, k = 3)
end_time <- Sys.time()
time_euc<-end_time-start_time 
paste(time_euc)
confusion_matrix_euc<-table(class_data_test,predict_knn_euclidian)
confusion_matrix_euc
accuracy_euc<-sum(diag(confusion_matrix_euc))/sum(confusion_matrix_euc)
accuracy_euc
```

In 8 by 8 confusion matrix, we can see true predicted instance counts. And accuracy of the model when k=3 is about 0.94 in Euclidian method.

```r
start_time <- Sys.time()
predict_knn_manhattan=kknn(class_data_train~., as.data.frame(combinedtraindata), as.data.frame(combinedtestdata),k = 1,distance = 1,kernel = "rectangular")
end_time <- Sys.time()
time_man<-end_time-start_time 
paste(time_man)
confusion_matrix_man<-table(class_data_test,predict_knn_manhattan$fitted.values)
confusion_matrix_man
accuracy_man<-sum(diag(confusion_matrix_man))/sum(confusion_matrix_man)
accuracy_man
```

In 8 by 8 confusion matrix, we can see true predicted instance counts. And accuracy of the model when k = 1 is about 0.95 in Euclidian method.
